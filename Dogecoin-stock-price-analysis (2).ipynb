```python
#Splitting the date column into day, month and year columns

dtf['Date']=pd.to_datetime(dtf['Date'])
dtf['Day']=dtf['Date'].dt.day
dtf['Month']=dtf['Date'].dt.month
dtf['Year']=dtf['Date'].dt.year

print(dtf)
```

```python
# Convert the 'Date' column to datetime format
dtf['Date'] = pd.to_datetime(dtf['Date'])

# Set 'Date' as the index for time-based plotting
dtf.set_index('Date', inplace=True)

# Time Series Plot - Dogecoin Prices Over Time
plt.figure(figsize=(12, 6))
plt.plot(dtf['Close'], label='Closing Price', color='blue')
plt.title('Dogecoin Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

**INFERENCE**

**1. Overall Trend: The plot reveals the overall trend in Dogecoin prices, showcasing periods of upward and downward movements.Understanding these trends is essential for investors and analysts seeking to make informed decisions.**

**2. Volatility Analysis: Fluctuations in the line reveal the volatility in Dogecoin prices. Sharp spikes or drops may indicate periods of increased market activity, speculative interest, or external factors influencing the cryptocurrency market.**

**3. Pattern Recognition: Repeated patterns or cycles in the plot may suggest recurring market behaviors. Identifying such patterns can aid in anticipating potential future price movements.**

```python
# Histogram - Distribution of Closing Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['Close'], bins=30, color='purple')
plt.title('Distribution of Dogecoin Closing Prices')
plt.xlabel('Closing Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

**INFERENCE**

**1. Dominant Price Ranges: The histogram reveals the most frequent closing price ranges that is 0.0, indicating where the market tends to concentrate.**

**2. Bimodal or Multimodal Patterns: The presence of multiple peaks in the histogram from 0.1 to 0.3 suggests distinct market regimes or phases with different dominant price levels.**

```python
# Moving Averages
df['Close_MA_20'] = df['Close'].rolling(window=20).mean()

plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Closing Price', color='blue')
plt.plot(df['Close_MA_20'], label='20-Day Moving Average', color='red')
plt.title('Dogecoin Prices with 20-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

**INFERENCE**

**1. In the plotted chart, the blue line represents the daily closing prices of Dogecoin, providing a clear view of the price fluctuations over time. The red line represents the 20-day moving average, which is a smoothed trend line calculated by taking the average of the closing prices over a 20-day period.**

**2. The moving average serves as a valuable indicator to identify trends, filter out short-term fluctuations, and highlight the underlying price direction**

```python
#Correlation Heatmap

correlation_matrix = dtf.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

**INFERENCE**ion.

**Strength and Direction of Correlation: The colors in the heatmap represent the strength and direction of the correlation between pairs of variables. Positive correlations are indicated by warmer colors (e.g., red), while negative correlations are represented by cooler colors (e.g., blue).**

**Correlation Coefficients: The numerical values within each cell of the heatmap represent the correlation coefficients. These values range from -1 to 1. A value closer to 1 indicates a strong positive correlation, while a value closer to -1 indicates a strong negative correlation. A value near 0 suggests a weak correlation.**

```python
# Boxplot of closing prices for each day of the month
plt.figure(figsize=(12, 6))
sns.boxplot(x='Day', y='Close', data=dtf, palette='viridis')
plt.title('Boxplot of Closing Prices for Each Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Closing Price (USD)')
plt.show()
     
```

**Inference:**

1. **Overall Spread:**
   The boxplot shows the overall spread and dispersion of closing prices for each day of the month. The length of the box represents the interquartile range (IQR), indicating where the central 50% of the data lies.

2. **Median Values:**
   The horizontal line within each box represents the median (50th percentile) closing price for the respective day. It provides a measure of central tendency for the distribution.

```python
# Plot average closing price by month
plt.figure(figsize=(12, 6))
average_closing_by_month = dtf.groupby('Month')['Close'].mean()
plt.bar(average_closing_by_month.index, average_closing_by_month.values, color='hotpink')
plt.title('Average Closing Price by Month')
plt.xlabel('Month')
plt.ylabel('Average Closing Price (USD)')
plt.show()
     
```

**Inference:**

1. **Monthly Trend Identification:**
   The plot allows for the identification of any noticeable trends or patterns in the average closing prices of Dogecoin across different months.

2. **Seasonal Patterns:**
   If there are consistent increases or decreases in average closing prices during specific months, it may suggest seasonal patterns or external factors influencing the cryptocurrency market.

3. **Comparison between Months:**
   Comparing the heights of the bars allows for a quick assessment of which months, on average, tend to have higher or lower closing prices. This information can be valuable for investors and traders.

```python
# Plot average closing price by year
plt.figure(figsize=(12, 6))
average_closing_by_year = dtf.groupby('Year')['Close'].mean()
plt.bar(average_closing_by_year.index, average_closing_by_year.values, color='coral')
plt.title('Average Closing Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Closing Price (USD)')
plt.show()
     
```

```python
# Overlaid Histogram of High and Low Prices
plt.figure(figsize=(12, 6))
plt.hist(dtf['High'], bins=20, color='yellow', edgecolor='black', alpha=0.7, label='High Prices')
plt.hist(dtf['Low'], bins=20, color='green', edgecolor='black', alpha=0.7, label='Low Prices')
plt.title('Overlaid Histogram of Dogecoin High and Low Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

**Inference:**

1. **Price Distribution:**
   The histogram shows the distribution of both high and low prices, allowing for a comparison of their respective ranges. The x-axis represents the price levels in USD, and the y-axis represents the frequency of occurrences.

2. **High-Low Spread:**
   Observing the spread between high and low prices provides insights into the price volatility within the given dataset. A wider spread may indicate periods of increased volatility, while a narrower spread suggests relative stability.

3. **Dominant Price Ranges:**
   Peaks in the histogram represent the most frequent price ranges for both high and low prices. Understanding these dominant ranges is crucial for identifying where the market tends to concentrate.

```python
# Horizontal Bar Plot of Mean High and Low Prices
mean_high = dtf['High'].mean()
mean_low = dtf['Low'].mean()

plt.figure(figsize=(10, 6))
plt.barh(['High Prices', 'Low Prices'], [mean_high, mean_low], color=['grey', 'mediumaquamarine'])
plt.title('Mean High and Low Prices of Dogecoin')
plt.xlabel('Mean Price (USD)')
plt.show()
```

**Inference:**

1. **Mean High and Low Prices:**
   The horizontal bars represent the mean (average) values for both high and low prices of Dogecoin. These values serve as measures of central tendency for each respective price category.

2. **Comparison between High and Low Prices:**
   The side-by-side comparison allows for a quick assessment of the average levels of high and low prices. Any notable differences in the lengths of the bars indicate disparities in the mean values between the two price categories.

3. **Relative Magnitudes:**
   Observing the lengths of the bars relative to each other provides a sense of the relative magnitudes of mean high and low prices. A longer bar indicates a higher mean value for that specific price category.

4. **Market Tendencies:**
